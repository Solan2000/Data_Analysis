```python
df=pd.read_csv('/mnt/HC_Volume_18315164/home-jupyter/jupyter-a-zakorko/3_taxi_peru.csv', sep=';',parse_dates=['start_at', 'end_at','arrived_at'])
ax=((df['driver_score'].value_counts(normalize=True))*100).round(2).reset_index()
ax=ax.rename(columns={'index':'driver_score','driver_score':'percentage'})
driver_score_counts=ax.sort_values('driver_score')
ax = sns.barplot(x='driver_score', y='percentage', data=driver_score_counts, color='blue', alpha=0.5)
ax.set(xlabel='Driver score', ylabel='Percentage')
sns.despine()  # убрать часть рамки графикаplt.show()
```

```python
df=pd.read_csv('/mnt/HC_Volume_18315164/home-jupyter/jupyter-a-zakorko/3_taxi_peru.csv', sep=';',parse_dates=['start_at', 'end_at','arrived_at'])
ax=((df['rider_score'].value_counts(normalize=True))*100).round(2).reset_index()
ax=ax.rename(columns={'index':'rider_score','rider_score':'percentage'})
rider_score_counts=ax.sort_values('rider_score')
ax = sns.barplot(x='rider_score', y='percentage', data=rider_score_counts, color='blue', alpha=0.5)
ax.set(xlabel='Rider score', ylabel='Percentage')
sns.despine()  # убрать часть рамки графикаplt.show()
```
