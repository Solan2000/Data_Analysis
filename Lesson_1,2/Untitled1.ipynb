```python
df=pd.read_csv('/mnt/HC_Volume_18315164/home-jupyter/jupyter-a-zakorko/2_bookings.csv',sep=';')
def format_column_names(df):
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    return df
format_column_names(df)
df['total_kids']=df.children+df.babies
df['has_kids']=df.total_kids!=0
churn_rate_with_kids = df[df['has_kids']]['is_canceled'].mean()*100
churn_rate_without_kids=df[~df['has_kids']]['is_canceled'].mean()*100
total=max(churn_rate_with_kids,churn_rate_without_kids).round(2)
total
```
